```python
subset = iot_train.iloc[:, :]
subset.plot()
#plt.plot(subset["DateTime"], subset["Vehicles"])
```

```python
decompose = seasonal_decompose(subset['Vehicles'], model='multiplicable', period=24)

residuals = decompose.resid
```

```python
#Creating DataFrame out of Residuals
df_resid = pd.DataFrame(residuals, subset.index)
df_resid = df_resid.dropna()
df_resid.plot()
plt.title("Vehicle Residuals Plot w Outliers")
```

```python
#DBSCAN on Vehicle Residuals
clustering = DBSCAN(eps=0.1).fit(df_resid)
labels = clustering.labels_
df_cluster = pd.DataFrame(labels, df_resid.index, columns=["labels"])
plt.scatter(df_resid.index, labels)
plt.title("DBSCAN Cluster of Residual Values from Vehicle Dataset")
```

```python
#Locating Anomalies
anomalies = df_cluster[df_cluster['labels']==-1]
anomalies
```

```python
df_resid['anomaly_score'] = [-1 if x in anomalies.index else 0 for x in df_resid.index]
anomalies = df_resid[df_resid['anomaly_score']==-1]
plt.plot(df_resid["resid"])
plt.scatter(df_resid.index, df_resid["resid"], facecolors="black")
plt.scatter(anomalies.index, anomalies["resid"], facecolor="Red")
plt.title("Vehicle Outlier Detection via DBSCAN")
```

```python
plt.plot(df_resid)
plt.scatter(df_resid.index, df_resid["resid"], facecolors="black")
plt.scatter(anomalies.index, anomalies["labels"], facecolor="Red")
plt.title("Passenger Data Outlier Detection vis DBSCAN")
```

```python
dfp = passengers
dfp["#Passengers"][10] = 300
dfp["#Passengers"][43] = 140
dfp["#Passengers"][100] = 550
dfp["#Passengers"][107] = 525
dfp["#Passengers"][127] = 300
```

```python
dfp.plot()
plt.title("Passengers w Outliers")
```

```python
decompose = seasonal_decompose(dfp['#Passengers'], model='multiplicable', period=12)

residuals = decompose.resid
```

```python
#Creating DataFrame out of Residuals
df_resid = pd.DataFrame(residuals, dfp.index)
df_resid = df_resid.dropna()
df_resid.plot()
plt.title("Passenger Residuals Plot w Outliers")
```

```python
#Z-Score Analysis
z_scores = (df_resid - np.mean(df_resid)) / np.std(df_resid)
plt.hist(z_scores)
plt.title("Z-Scores of Passenger Residuals")
```

```python
zs = stats.zscore(df_resid['resid'])
df_resid['zscores'] = zs
outliers = df_resid[abs(df_resid["zscores"]) >= 2]
outliers
```

```python
#DBSCAN on Passenger Residuals
clustering = DBSCAN(eps=0.1).fit(df_resid)
labels = clustering.labels_
df_cluster = pd.DataFrame(labels, df_resid.index, columns=["labels"])
plt.scatter(df_resid.index, labels)
plt.title("DBSCAN Cluster of Residual Values from Passenger Dataset")
```

```python
#Locating Anomalies
anomalies = df_cluster[df_cluster['labels']==-1]
anomalies
```

```python
plt.plot(df_resid)
plt.scatter(df_resid.index, df_resid["resid"], facecolors="black")
plt.scatter(anomalies.index, anomalies["labels"], facecolor="Red")
plt.title("Passenger Data Outlier Detection vis DBSCAN")
```

```python
#Outlier Detection on Passengers via Isolation Forests
model = IsolationForest(contamination=0.05)
model.fit(df_resid)
df_resid['scores'] = model.decision_function(df_resid[['resid']])
df_resid['anomaly_score'] = model.predict(df_resid[['resid']])
anomalies = df_resid[df_resid['anomaly_score']==-1]
anomalies.head()
```

```python
plt.plot(df_resid["resid"])
plt.scatter(df_resid.index, df_resid["resid"], facecolors="black")
plt.scatter(anomalies.index, anomalies["resid"], facecolor="Red")
plt.title("Passenger Data Outlier Detection via Isolation Forests")
```
